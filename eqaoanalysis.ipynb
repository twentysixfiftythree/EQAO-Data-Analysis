# EQAO Schools Analysis

## Introduction
This notebook provides an exploratory data analysis (EDA) of the EQAO dataset containing information about schools, their sizes, towns, and EQAO achievement scores.

## Data Loading and Cleaning



```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r'C:\Users\isaac\Documents\EQAODataAnalysis\EQAODataset(onlymatches).csv')
data.head()
```

### Convert "School Size" column to integer format

```python
data['School Size'] = data['School Size'].str.replace(',', '').astype(int)
data.head()
```

## Exploratory Data Analysis (EDA)
### Descriptive Statistics

```python
data.describe()
```

## Distribution of EQAO Achievement Scores

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))
sns.histplot(data=data, x='Applied EQAO Achievement', kde=True, ax=ax[0], color="skyblue")
sns.histplot(data=data, x='Academic EQAO Achievement', kde=True, ax=ax[1], color="salmon")

ax[0].set_title('Distribution of Applied EQAO Achievement')
ax[1].set_title('Distribution of Academic EQAO Achievement')

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(10, 10))
sns.scatterplot(data=data, x='School Size', y='Applied EQAO Achievement', ax=ax[0], color="skyblue", alpha=0.6)
sns.scatterplot(data=data, x='School Size', y='Academic EQAO Achievement', ax=ax[1], color="salmon", alpha=0.6)

ax[0].set_title('School Size vs. Applied EQAO Achievement')
ax[1].set_title('School Size vs. Academic EQAO Achievement')

plt.tight_layout()
plt.show()
```

```python
# Create a figure and a set of subplots without School Name labels on the x-axis
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(15, 12))

# Box plot for Applied EQAO Achievement across different towns without x-axis labels
sns.boxplot(data=data, x='School Town', y='Applied EQAO Achievement', ax=ax[0], palette="Blues")
ax[0].set_title('Variation of Applied EQAO Achievement Across Towns')
ax[0].set_xlabel('')
ax[0].set_xticks([])  # Remove x-axis ticks

# Box plot for Academic EQAO Achievement across different towns without x-axis labels
sns.boxplot(data=data, x='School Town', y='Academic EQAO Achievement', ax=ax[1], palette="Reds")
ax[1].set_title('Variation of Academic EQAO Achievement Across Towns')
ax[1].set_xlabel('')
ax[1].set_xticks([])  # Remove x-axis ticks

# Adjust the layout
plt.tight_layout()
plt.show()
```

### Variation of Applied EQAO Achievement Across Towns:
There is noticeable variability in the applied EQAO achievement across different towns. Some towns, like Ajax and Aurora, tend to have lower median achievement rates, while others, such as Belleville, have higher median rates.
The spread of the data (as shown by the interquartile range) is also different across towns. Some towns have a tighter distribution of scores, while others have a wider spread.

### Variation of Academic EQAO Achievement Across Towns:
The academic EQAO achievement also varies across towns, but the differences are not as pronounced as in the applied achievement rates. Most towns have a median achievement rate above 0.8.
Some towns, like Belleville, have very tight distributions, suggesting consistency in academic achievement across schools in those towns.

## Conclusions:

 Size and Achievement: There doesn't seem to be a strong linear correlation between school size and either applied or academic EQAO achievement. Schools of all sizes have a range of achievement scores.

Town-wise Achievement: There is variability in EQAO achievement scores across different towns. While some towns consistently perform better than others, it's essential to consider other factors like resources, socio-economic conditions, and more when interpreting these results.
